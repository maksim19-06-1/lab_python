```python
def prime_number_2(q):
    z = 2
    while (z<=q):
        if((q%z) == 0):
            if(q == z):
                return q 
            else:
                z = q+1
        else:
            z+=1
    return 0
def prime_number(i):
    m = 2
    x = 2
    w = 0
    while (x<=m):
        if(prime_number_2(x)!=0):
            w+=1
            if(w == i):
                return x
        x+=1
        m+=1
    return 0 
def is_number(s):
    b = ['5', '2', '34', 'ff'] 
    try:
        t = int(b[s])
        print(t,'- те просте число - це',prime_number(t))
    except ValueError:
        print('Помилка. Присутнє нечислове значення!')
    return 0
i=0
for i in range(4):
    is_number(i)
    i+=1
```
